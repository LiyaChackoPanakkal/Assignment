# 3. Plot a bar graph of male vs female buyers participated in the sales

```python
import matplotlib.pyplot as plt
Gender=data['Buyer Gender'].value_counts().to_frame().reset_index()
fig = plt.figure(figsize = (8, 8))
plt.bar(Gender['index'],Gender['Buyer Gender'],color=['red','blue'],width = 0.25)
plt.title('Female vs Male Participents')
plt.xlabel('Gender')
plt.ylabel('Count')
```

#### Insights:Compared to male participants, the number of female participants is slightly higher in the sale
