# Case Study on ANOVA


### XYZ Company has offices in four different zones. The company wishes to investigate the following :


```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv('Sales_data_zone_wise.csv')
data.head()
```

### ● The mean sales generated by each zone.


```python
np.mean(data)
```

### ● Total sales generated by all the zones for each month.


```python
d=data.set_index("Month")
d.head()
```

```python
d.sum(axis=1)
```

### ● Check whether all the zones generate the same amount of sales.


```python
from scipy import stats

fvalue,pvalue=stats.f_oneway(data["Zone - A"],data["Zone - B"],data["Zone - C"],data["Zone - D"])
print(fvalue,pvalue)
```

```python
pvalue = 0.0011827601694503335
print('pvalue =',pvalue)
if pvalue < 0.05:
    print('Reject the null hypothesis(Ho)')
else:
    print('Accept null hypothesis(Ho)')
```

####  Insights:  Since p value is less than alpha value(0.05), we can reject null hypothesis. Hence,All the zones generate the different amount of sales.
