# 3. Plot a bar graph of male vs female buyers participated in the sales.

```python
new1 = data["Buyer Gender"].isin(["Male"])
#number of male
n1=sum(new1)
n1           
```

```python
new2 = data["Buyer Gender"].isin(["Female"])
#number of female
n2=sum(new2)
n2
```

```python
import matplotlib.pyplot as plt
fig = plt.figure()
dataset = {n1:'Male',n2:'Female'}
Gender = list(dataset.values())
values = list(dataset.keys())
fig = plt.figure(figsize = (8, 8))
plt.bar(Gender, values, color ='red',
        width = 0.25)
plt.xlabel("Gender")
plt.ylabel("No. of Participents in sail")
plt.title("Female vs Male Participents")
plt.show()
```
