# Perform meaningful visualizations using the dataset.

## #Scatter Plot

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(8,8))
plt.scatter(data['SL'],data['SW'],c='red',s=12)
plt.title('Sepal Length & Petal Length')
plt.xlabel('SL')
plt.ylabel('PL')
```

# # Pairplot

```python
sns.pairplot(data)
```

```python
slice_data=data[['SL','PL','PW']]
sns.pairplot(slice_data)
```

# #Strip Plot

```python
sns.stripplot(x='SW',y='PW',data=data)
plt.title('Sepal Width & Petal Width')
```
